### Importación de datos



```python
import pandas as pd
import matplotlib as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
ingreso_total_tienda1 = tienda['Precio'].sum()
print(f'El ingreso de la tienda 1 es {ingreso_total_tienda1}')
```

```python
ingreso_total_tienda2 = tienda2['Precio'].sum()
print(f'El ingreso de la tienda 1 es {ingreso_total_tienda2}')
```

```python
ingreso_total_tienda3 = tienda3['Precio'].sum()
print(f'El ingreso de la tienda 1 es {ingreso_total_tienda3}')
```

```python
ingreso_total_tienda4 = tienda4['Precio'].sum()
print(f'El ingreso de la tienda 1 es {ingreso_total_tienda4}')
```

```python
def ingresos_totales_tiendas():
  tiendas = [ingreso_total_tienda1,ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
  etiquetas_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

  plt.ticklabel_format(style='plain', axis='y')

  plt.ylim(0,max(tiendas)*1.05)

  plt.bar(etiquetas_tiendas, tiendas, color=['Purple', 'Red', 'Grey', 'Blue'], label='Ingresos')


  plt.title('Ingresos Totales por Tienda')
  plt.xlabel('Tiendas')
  plt.legend()
  plt.ylabel('Ingresos en $')
ingresos_totales_tiendas()
```

# 3. Calificación promedio de la tienda


```python
suma_calificaciones = tienda['Calificación'].sum()

suma_n_calificaciones = tienda['Calificación'].count()

promedio_calificaciones_tienda1 = round((suma_calificaciones / suma_n_calificaciones),2)

print(f'El promedio de calificaciones de la tienda 1 es {promedio_calificaciones_tienda1}')
```

```python
suma_calificaciones = tienda2['Calificación'].sum()

suma_n_calificaciones = tienda2['Calificación'].count()

promedio_calificaciones_tienda2 = round((suma_calificaciones / suma_n_calificaciones),2)

print(f'El promedio de calificaciones de la tienda 1 es {promedio_calificaciones_tienda2}')
```

```python
suma_calificaciones = tienda3['Calificación'].sum()

suma_n_calificaciones = tienda3['Calificación'].count()

promedio_calificaciones_tienda3 = round((suma_calificaciones / suma_n_calificaciones),2)

print(f'El promedio de calificaciones de la tienda 1 es {promedio_calificaciones_tienda3}')
```

```python
suma_calificaciones = tienda4['Calificación'].sum()

suma_n_calificaciones = tienda4['Calificación'].count()

promedio_calificaciones_tienda4 = round((suma_calificaciones / suma_n_calificaciones),2)

print(f'El promedio de calificaciones de la tienda 1 es {promedio_calificaciones_tienda4}')
```

```python
def calificaciones_tiendas():
  tiendas = [promedio_calificaciones_tienda1, promedio_calificaciones_tienda2, promedio_calificaciones_tienda3, promedio_calificaciones_tienda4]
  etiquetas_tienda = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
  plt.xlabel('Calificación')
  plt.ylabel('Tiendas')
  plt.title('Calificación por cada tienda')
  barras = plt.barh(etiquetas_tienda, tiendas, color=['#FF5F42', '#FAB78C', '#00C105', '#0092FA'])
  plt.xlim(right=4.5)

  for barra in barras:
    ancho = barra.get_width()

    centro_barra = barra.get_y() + barra.get_height() / 2

    plt.text(
        ancho + 0.05,
        centro_barra,
        f'{ancho:.2f}',
        va='center',
        ha='left'

    )
```

# 5. Envío promedio por tienda

```python
promedio_envio_tienda1 = round(tienda['Costo de envío'].mean(),2)
print(f'El promedio de la tienda 1 es: {promedio_envio_tienda1}')
```

```python
promedio_envio_tienda2 = round(tienda2['Costo de envío'].mean(),2)
print(f'El promedio de la tienda 1 es: {promedio_envio_tienda2}')
```

```python
promedio_envio_tienda3 = round(tienda3['Costo de envío'].mean(),2)
print(f'El promedio de la tienda 1 es: {promedio_envio_tienda3}')
```

```python
promedio_envio_tienda4 = round(tienda4['Costo de envío'].mean(),2)
print(f'El promedio de la tienda 1 es: {promedio_envio_tienda4}')
```

```python
def promedio_envios():
  tiendas_envios = [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]
  tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
  plt.grid()
  plt.title('Costo envio promedio por tienda')
  plt.ylabel('Costo del envio')
  plt.plot(tiendas, tiendas_envios, marker='o', linestyle='--', color='Green', linewidth=2,markeredgewidth=3,label='Tendencia')
  plt.legend()
```

# **Informe Final**

## **Introducción**
El propósito de este informe es ayudar al Sr. Juan a tomar una decisión estratégica sobre su negocio. El objetivo es identificar cuál de las cuatro tiendas actuales no se debería quedar y, por lo tanto, debería venderse para optimizar recursos y no perder dinero.

Para ello, he realizado un análisis de datos comparativo, evaluando cada tienda según métricas clave:

*   Ingresos totales
*   Calificación promedio de clientes
*   Costo promedio de envío
*   Ventas por categoría
*   Productos más y menos vendidos


## **Desarrollo**

El análisis revela un panorama claro. Si bien todas las tiendas tienen fortalezas y debilidades, una de ellas muestra un rendimiento financiero significativamente inferior que justifica su venta.

1. El Factor Decisivo: Ingresos Totales
El indicador financiero más importante es el ingreso total que genera cada tienda. Como se observa en el siguiente gráfico, las tiendas 1, 2 y 3 mantienen un nivel de ingresos comparable y robusto. Sin embargo, la Tienda 4 se sitúa notablemente por debajo de las demás.

```python
ingresos_totales_tiendas()
```

2. Contraste de Métricas: Operaciones vs. Finanzas
A primera vista, la Tienda 4 parece una buena candidata. El análisis de la experiencia del cliente muestra que tiene la calificación promedio más alta (4.05) y, de forma destacada, el costo de envío promedio más bajo.

```python
calificaciones_tiendas()
```

```python
promedio_envios()
```

Sin embargo, el punto clave del análisis es que estas ventajas operativas no se están traduciendo en un rendimiento financiero. A pesar de tener clientes aparentemente satisfechos y la logística de envío más barata, la tienda no logra generar un volumen de ingresos equiparable al de sus pares.

## **Conclusión y Recomendación**

Basado en el análisis completo, la recomendación es proceder con la venta de la Tienda 4.

Aunque la Tienda 4 demuestra fortalezas en la experiencia del cliente (alta calificación y bajo costo de envío), su bajo rendimiento en Ingresos Totales la posiciona como la operación menos rentable del grupo.

Mantener esta tienda genera un gasto innecesario en manutención que podría ser reinvertido de manera mucho más efectiva en las tiendas 1, 2 y 3, las cuales demuestran un mayor y más saludable potencial de ingresos.
